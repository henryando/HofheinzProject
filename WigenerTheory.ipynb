```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

Let's start actually plotting some of the figures from the paper. As a first test, we ignore any decoherencee or other time dependent effects, and only consider the theoretical Wigener tomographies of the prepared states. First, let's create a parity operator

```python
def parity(N):
    par = np.identity(N)
    i = 0
    while i < N:
        par[i,i] = -1
        i = i + 2
    return Qobj(inpt=par)
```

Using the parity operator, we can create a Wigener tomography operator

```python
def wig(N,a):
    return displace(N,-a).dag()*parity(N)*displace(N,-a)
    
```

We apply the Wigener tomography operator to resonator states $|0\rangle + |1\rangle$,$|0\rangle + |2\rangle$,
$|0\rangle + |3\rangle$,$|0\rangle + |4\rangle$, and $|0\rangle + |5\rangle$ with values of $\alpha$ for
Re($\alpha$) over [-2,2] and Im($\alpha$) also over [-2,2]

```python
np.random.seed(0)
import seaborn as sns
sns.set()

N = 30
fig = plt.figure()

T = 100
Im_a = np.linspace(-2, 2, T)
Re_a = np.linspace(-2, 2, T)
output = np.zeros([T,T,5])

for k in range(0,5):
    psi0 = (1/np.sqrt(2))*(fock(N, k+1) + fock(N, 0))
    for i in range(0,T):
        if i%1 ==0:
            clear_output(wait=True)
            print(i/T/5+k/5)
        for j in range(0,T):
            output[j,i,k] = expect(wig(N,np.complex(Re_a[i],Im_a[j])),psi0)
 
```

```python
np.random.seed(0)
import seaborn as sns
sns.set()

ax = sns.heatmap(output[:,:,0],center = 0,xticklabels=np.round(Re_a,2), yticklabels=np.round(Im_a,2), 
                 cmap=sns.diverging_palette(250,10,  s=90, l=60,n=1000, center="dark"))
for ind, label in enumerate(ax.get_xticklabels()):
    if ind % 20 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
for ind, label in enumerate(ax.get_yticklabels()):
    if ind % 20 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
ax.set_title('|0> + |1>')
ax.set(xlabel='Re(a)', ylabel='Im(a)')
```
